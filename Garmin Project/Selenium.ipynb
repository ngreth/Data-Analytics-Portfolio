# Learning Selenium

```python
# Import relevant libraries

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# id, name, and class name are the most common ways to identify an element
# id is unique
# name is usually unique
# class is okay to search by, but class is not a unique identifier - selenium will return the first outcome

s=Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=s)
driver.maximize_window()

driver.get("https://techwithtim.net/")
search = driver.find_element("name", "s")
search.send_keys("test")
search.send_keys(Keys.RETURN)

# Allow the user 5 seconds to see the outcome
## time.sleep (5)

# Explicit waits - to make sure values load prior to selenium moving onto the next step
try:
    main = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "main"))
    )
    print(main.text)
    
    articles = main.find_elements(By.TAG_NAME,"article")
    for article in articles:
        header = article.find_element(By.CLASS_NAME,"entry-summary")
        #print(header.text)
        print(header)
        
    
finally:
    driver.quit()
```
