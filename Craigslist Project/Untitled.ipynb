```python
# Importing the required libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# Downloading contents of the web page
url = "https://en.wikipedia.org/wiki/List_of_manufacturers_by_motor_vehicle_production"
data = requests.get(url).text
```

```python
# Creating BeautifulSoup object
soup = BeautifulSoup(data, 'html.parser')
```

```python
# Verifying tables and their classes
print('Classes of each table:')
for table in soup.find_all('table'):
    print(table.get('class'))
```

```python
table = soup.find('table', class_='sortable')
table
```

```python
# Collecting Ddata
rank = []
group = []
country = []
vehicles = []
for row in table.tbody.find_all('tr'):    
    # Find all data for each column
    columns = row.find_all('td')
#    print(columns)
    if(columns != []):
        rank.append(columns[0].text.strip())
        group.append(columns[1].text.strip())
        country.append(columns[2].text.strip())
        vehicles.append(columns[3].text.strip())
#print(group)
# Defining of the dataframe
import numpy as np
d = {'rank': rank, 'group': group, 'country': country, 'vehicles': vehicles}
df = pd.DataFrame(data=d, index = np.arange(0,len(rank),1))
```
