```python
# Import libraries
from bs4 import BeautifulSoup as bs
import requests
import smtplib
import time
import datetime
import numpy as np
import pandas as pd
import html5lib
```

```python
# Load webpage content
r = requests.get("https://austin.craigslist.org/search/austin-tx/cta?lat=30.26990&lon=-97.71600&search_distance=1.6")

# Convert webpage content to a beautiful soup object
soup = bs(r.content)

# Print with indentations, so the webpage content is easier to read
print(soup.prettify())
```

```python
title = soup.find_all('h3')
print(title[119].get_text())

titles = []

for i in np.arange(0,len(title)):
    titles.append(title[i].get_text())
    
titles
```
